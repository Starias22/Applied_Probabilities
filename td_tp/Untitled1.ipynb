```python
import numpy as np
import matplotlib.pyplot as plt
 
n=10000
h=10**-2
X=np.zeros((n,))
def func():
    U=np.random.uniform(0,1)
    V=np.random.uniform(0,1)
    return U*V
for i in range(n):
    X[i]=func()
    
#print(X)
def f(t):
    cpt=0
    for i in range(n):
        if (X[i] >= t) and (X[i] <= t+h):
            cpt+=1
    return cpt/(n*h)

def F(t):
    cpt=0
    for i in range(n):
        if (X[i] <= t):
            cpt+=1
    return cpt/n

Xcoords=np.linspace(start=0.1,stop=1,num=1000)
#print(P)
Ycoords=[f(t) for t in Xcoords]
plt.plot(Xcoords,Ycoords,color='red',label='fX')
Ycoords=[F(t) for t in Xcoords]
plt.plot(Xcoords,Ycoords,color='magenta',label='FX')
plt.legend(loc='upper left')
plt.title('Approximation of f and F')
```

```python
Ycoords=[-np.log(t) for t in Xcoords]
plt.plot(Xcoords,Ycoords,color='red',label='fX')
Ycoords=[t-t*np.log(t) for t in Xcoords]
plt.plot(Xcoords,Ycoords,color='magenta',label='FX')
plt.legend(loc='upper left')
plt.title('f and F')
```
